# 阶数阶乘(泰勒展开)

```python
# 求函数f(x)=e**x的n阶麦克林展开式,n=10(画出变化率)
from sympy import *
```

```python
x = symbols('x')
Fn = E**x # 目标函数
fn = series(Fn,x,n=10) # 泰勒展开方法
fn
```

```python
# 变化率
plotting.plot(fn.subs(O(x**10),0),(x,0,10))  # 麦克劳林展开式 
plotting.plot(Fn,(x,0,10)) # 原函数  
```

```python
# 求函数f(x)=sinx的n阶麦克林展开式,n=4,6,8,10,（画出原图与不同展开式的变化率)
x = symbols('x')
Fn = sin(x)
fn = series(Fn,x,n=8)
fn
plotting.plot(fn.subs(O(x**8),0),(x,0,4))
plotting.plot(Fn,(x,0,4))
```
