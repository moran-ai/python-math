```python
# 求z = x**+3*x*y+y**2在点(1,2)的偏导数
from sympy import *
x,y = symbols("x,y")
fn=x**2+3*x*y+y**2
fx = diff(fn,x)
fy = diff(fn,y)  # 求偏导数
fx.subs({x:1,y:2})  # 求函数值subs
```

```python
fy.subs({x:1,y:2}).evalf(2)  # 浮点数evalf
```

```python
# 方向导数,求f(x,y,z)在点(1,1,2)沿方向L的方向导数,其中L的方向角分别为60,45,60
from sympy import * 
x,y,z = symbols('x,y,z')
fn = x*y+y*z+z*x
fx = diff(fn,x)  # 先求偏导数
fy = diff(fn,y)
fz = diff(fn,z)
fp = fx*cos(pi/3)+fy*cos(pi/4)+fz*cos(pi/3) # 方向余弦*每一项偏导数
fp
# fp.subs({x:1,y:1,z:2}).evalf(5)
```

```python
# 求梯度
from sympy.vector import * # 这是一个向量工具库
C = CoordSys3D('C') # 选取坐标轴
f = 1/2*(C.x**2+C.y**2)
gradient(f)
```

```python
res = gradient(f) # 求梯度
res.subs({C.x:1,C.y:1})
```

```python
res.normalize() # 求向量方向
```

```python
res.magnitude() # 求向量的模,梯度的模及最大方向导数.
```
