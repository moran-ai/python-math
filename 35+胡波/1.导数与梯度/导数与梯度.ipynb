```python
# z = x**+3*x*y+y**2(1,2)
from sympy import *
x,y = symbols("x,y")
fn=x**2+3*x*y+y**2
fx = diff(fn,x)
fy = diff(fn,y)  # 
fx.subs({x:1,y:2})  # subs
```

```python
fy.subs({x:1,y:2}).evalf(2)  # evalf
```

```python
# 
from sympy.vector import * # 
C = CoordSys3D('C') # 
f = 1/2*(C.x**2+C.y**2)
gradient(f)
```

```python
res = gradient(f) # 
res.subs({C.x:1,C.y:1})
```

```python
res.normalize() # 
```
